```python
from socket import*
import time

HOST = '127.0.0.1'
PORT = 50000
MAXSIZE = 50000

cntSock = socket(AF_INET, SOCK_DGRAM)

cntSock.connect((HOST, PORT))

print("서버 연결 완료.")
msg = input("입력: ")

start = time.time()
cntSock.sendto(msg.encode(),(HOST, PORT))

data = cntSock.recv(1024)
total_messages = int(data.decode())

data = b''
print(total_messages)
for i in range(total_messages):
    packet = cntSock.recv(MAXSIZE + 1024)
    print(f"{i}번째 패킷 수신.")
    data += packet

end = time.time()
print("검사 시작.")
cnt = data.count(b'A')

print(f"문자열 크기: {msg.split(' ')[1]} : {cnt}")
print(f"UDP 송수신 시간: {end - start:.5f} sec")

cntSock.close()
```
